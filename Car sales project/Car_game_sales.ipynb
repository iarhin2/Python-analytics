```python
df = df.dropna(subset=['Year', 'Publisher'])
```

```python
df.isnull().sum()
```

```python
# Remove duplicate records
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.columns
```

```python
# Pivot tables using groupby()
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)
```

```python
# Top 10 publishers by sales
pivot1 = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
# Visualization
# publisher by sales
pivot1.plot(kind='bar')
plt.title('Top 10 Publishers by Sales')
plt.xlabel('Publisher')
plt.ylabel('Sales (millions)')
plt.show()
```

```python
# Genre with the most sales in North America
df.groupby('Genre')['NA_Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('Genre')['NA_Sales'].sum().plot(kind='bar')
plt.title('Genre with the most sales in North America')
plt.xlabel('Genre')
plt.ylabel('Sales (millions)')
plt.show()
```
